## KNN 실습 p.241

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
import matplotlib.pyplot as plt
```

### 1. 데이터 구성하기

```python
ds_tc = pd.read_csv("실습파일/통신고객이탈.csv")
ds_tc.head()
```

```python
ds_tc["CHURN"].replace(to_replace = "Active", value = 0, inplace = True)
ds_tc["CHURN"].replace(to_replace = "Churned", value = 1, inplace = True)
```

```python
ds_tc.isnull().sum(axis=0)
```

```python
ds_tc_dummy = pd.get_dummies(ds_tc.drop(["CHURN", "CUSTOMER_ID"], axis =1, inplace = False))
ds_tc_dummy.head()
```

```python
from sklearn.model_selection import train_test_split


ds_tc_y = ds_tc["CHURN"]
ds_tc_x = ds_tc_dummy

ds_tc_train_x, ds_tc_test_x, ds_tc_train_y, ds_tc_test_y =\
train_test_split(ds_tc_x, ds_tc_y, test_size=0.30, random_state = 1234)

print("train data X size:{}".format(ds_tc_train_x.shape))
print("train data Y size:{}".format(ds_tc_train_y.shape))
print("test data X size:{}".format(ds_tc_test_x.shape))
print("test data Y size:{}".format(ds_tc_test_y.shape))
```

```python
ds_tc_dummy.var(axis=0)
```

- 변수들의 분산 차가 크기 때문에 스케일링하여 분석하기로 결정한다.

```python
from sklearn.preprocessing import scale

ds_tc_train_x_scaled = scale(ds_tc_train_x,axis = 0)
ds_tc_test_x_scaled = scale(ds_tc_test_x,axis = 0)
pd.DataFrame(ds_tc_train_x_scaled, columns=ds_tc_train_x.columns).head()
```

### 2. 모델 생성

```python
knn_uncustomized = KNeighborsClassifier()
knn_uncustomized
```

```python
knn_uncustomized.fit(ds_tc_train_x_scaled, ds_tc_train_y)

y_pred = knn_uncustomized.predict(ds_tc_test_x_scaled)
print("predict test set: {}".format(y_pred.tolist()[:10]))
print("test set label: {}".format(ds_tc_test_y.tolist()[:10]))
print("accuracy: {0: .3f}".format(knn_uncustomized.score(ds_tc_test_x_scaled,ds_tc_test_y)))
```

```python
max_n_neighbors = 10
para_n_neighbors = [i+1 for i in range(max_n_neighbors)]
train_accuracy=[]
test_accuracy=[]

for n_neighbors in para_n_neighbors:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(ds_tc_train_x_scaled, ds_tc_train_y)
    train_accuracy.append(clf.score(ds_tc_train_x_scaled, ds_tc_train_y))
    test_accuracy.append(clf.score(ds_tc_test_x_scaled, ds_tc_test_y))
    
ds_neighbors = pd.DataFrame()
ds_neighbors["Neighbors"] = para_n_neighbors
ds_neighbors["TrainAccuracy"] = train_accuracy
ds_neighbors["TestAccuracy"] = test_accuracy
ds_neighbors
```

```python
ds_neighbors.set_index("Neighbors", inplace = False).plot()
```

```python
max_n_neighbors = 10
para_n_neighbors = [i+1 for i in range(max_n_neighbors)]*2
para_weights = ["uniform"]*max_n_neighbors + ["distance"]*max_n_neighbors
train_accuracy=[]
test_accuracy=[]

for (n_neighbors, weights) in zip(para_n_neighbors, para_weights):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights)
    clf.fit(ds_tc_train_x, ds_tc_train_y)
    train_accuracy.append(clf.score(ds_tc_train_x, ds_tc_train_y))
    test_accuracy.append(clf.score(ds_tc_test_x, ds_tc_test_y))
    
ds_neighbors_weight = pd.DataFrame()
ds_neighbors_weight["Neighbors"] = para_n_neighbors
ds_neighbors_weight["Weights"] = para_weights
ds_neighbors_weight["TrainAccuracy"] = train_accuracy
ds_neighbors_weight["TestAccuracy"] = test_accuracy
ds_neighbors_weight
```

```python
ds_neighbors_weight_melt = pd.melt(ds_neighbors_weight, id_vars=["Neighbors", "Weights"])

ds_neighbors_weight_melt["Accuracy"] = ds_neighbors_weight_melt["Weights"] + "_"+ ds_neighbors_weight_melt["variable"]

ax = sns.lineplot(x="Neighbors", y ="value",hue= "Accuracy", data=ds_neighbors_weight_melt)
```

- 가중치가 distance 일 때는 train data 정확도는 1
- 가중치는 Uniform을 선택하기로 결정

```python
max_n_neighbors = 10
para_n_neighbors = [i+1 for i in range(max_n_neighbors)]*3
para_algorithm = ["ball_tree"]*max_n_neighbors + ["kd_tree"]*max_n_neighbors + ["brute"]*max_n_neighbors
train_accuracy=[]
test_accuracy=[]

for (n_neighbors, algorithm) in zip(para_n_neighbors, para_algorithm):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors, weights='uniform', algorithm=algorithm)
    clf.fit(ds_tc_train_x, ds_tc_train_y)
    train_accuracy.append(clf.score(ds_tc_train_x, ds_tc_train_y))
    test_accuracy.append(clf.score(ds_tc_test_x, ds_tc_test_y))
    
ds_neighbors_algorithm = pd.DataFrame()
ds_neighbors_algorithm["Neighbors"] = para_n_neighbors
ds_neighbors_algorithm["Algorithm"] = para_algorithm
ds_neighbors_algorithm["TrainAccuracy"] = train_accuracy
ds_neighbors_algorithm["TestAccuracy"] = test_accuracy
ds_neighbors_algorithm
```

```python
ds_neighbors_algorithm_melt = pd.melt(ds_neighbors_algorithm, id_vars=["Neighbors", "Algorithm"])

ds_neighbors_algorithm_melt["Accuracy"] = ds_neighbors_algorithm_melt["Algorithm"] + "_"+ ds_neighbors_algorithm_melt["variable"]

ax = sns.lineplot(x="Neighbors", y ="value",hue= "Accuracy", data=ds_neighbors_algorithm_melt)
```

- 알고리즘 별 정확도의 차이가 없음
- 알고리즘은 ball_tree알고리즘을 선택하기로 결정

```python
max_n_neighbors = 10
para_n_neighbors = [i+1 for i in range(max_n_neighbors)]*3
para_metric = ["minkowski"]*max_n_neighbors + ["euclidean"]*max_n_neighbors + ["manhattan"]*max_n_neighbors
train_accuracy=[]
test_accuracy=[]

for (n_neighbors, metric) in zip(para_n_neighbors, para_metric):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors, weights='uniform', algorithm="ball_tree", metric= metric)
    clf.fit(ds_tc_train_x, ds_tc_train_y)
    train_accuracy.append(clf.score(ds_tc_train_x, ds_tc_train_y))
    test_accuracy.append(clf.score(ds_tc_test_x, ds_tc_test_y))
    
ds_neighbors_metric = pd.DataFrame()
ds_neighbors_metric["Neighbors"] = para_n_neighbors
ds_neighbors_metric["Metric"] = para_metric
ds_neighbors_metric["TrainAccuracy"] = train_accuracy
ds_neighbors_metric["TestAccuracy"] = test_accuracy
ds_neighbors_metric
```

```python
ds_neighbors_metric_melt = pd.melt(ds_neighbors_metric, id_vars=["Neighbors", "Metric"])

ds_neighbors_metric_melt["Accuracy"] = ds_neighbors_metric_melt["Metric"] + "_"+ ds_neighbors_metric_melt["variable"]

ax = sns.lineplot(x="Neighbors", y ="value",hue= "Accuracy", data=ds_neighbors_metric_melt)
```

### 3. 결론 도출

- metric 별 정확도의 차이가 거의 없다 
- 거리계산 metric은 manhattan 을 사용하기로 결정

```python
clf_model = KNeighborsClassifier(n_neighbors=6, weights='uniform', algorithm='ball_tree', metric='manhattan')
clf_model.fit(ds_tc_train_x, ds_tc_train_y)
print("train data accuracy: {0: .3f}".format(clf_model.score(ds_tc_train_x,ds_tc_train_y)))
print("test data accuracy: {0: .3f}".format(clf_model.score(ds_tc_test_x, ds_tc_test_y)))
```

- 예측 정확도 : 0.842
- 이웃 수 : 6
- 가중치 : 균일(uniform)
- 알고리즘 : 차이없음
- 거리 계산: manhattan

## KNN 숙제

### 1. 데이터 불러오기

```python
df = pd.read_csv("실습파일/유방암.csv", engine = 'python')
df.head()
```

```python
df_dummy = pd.get_dummies(df)
df_dummy.head()
```

```python
from sklearn.model_selection import train_test_split


df_y = df["diagnosis"]
df_x = df_dummy.drop("diagnosis", axis = 1, inplace = False)

df_train_x, df_test_x, df_train_y, df_test_y =\
train_test_split(df_x, df_y, test_size=0.30, random_state = 1234)


print("train data X size:{}".format(df_train_x.shape))
print("train data Y size:{}".format(df_train_y.shape))
print("test data X size:{}".format(df_test_x.shape))
print("test data Y size:{}".format(df_test_y.shape))
```

```python
df_train_x_scaled = scale(df_train_x,axis = 0)
df_test_x_scaled = scale(df_test_x,axis = 0)
pd.DataFrame(df_train_x_scaled, columns=df_train_x.columns).head()
```

### 2. 모델 생성하기

```python
knn_uncustomized = KNeighborsClassifier()
knn_uncustomized
```

```python
knn_uncustomized.fit(df_train_x, df_train_y)

y_pred = knn_uncustomized.predict(df_test_x)
print("predict test set: {}".format(y_pred.tolist()[:10]))
print("test set label: {}".format(df_test_y.tolist()[:10]))
print("accuracy: {0: .3f}".format(knn_uncustomized.score(df_test_x,df_test_y)))
```

```python
knn_uncustomized.fit(df_train_x_scaled, df_train_y)

y_pred = knn_uncustomized.predict(df_test_x_scaled)
print("predict test set: {}".format(y_pred.tolist()[:10]))
print("test set label: {}".format(df_test_y.tolist()[:10]))
print("accuracy: {0: .3f}".format(knn_uncustomized.score(df_test_x_scaled,df_test_y)))
```

- scale 했을 때 정확도가 더 높으므로 스케일한 데이터를 분석에 사용한다.

```python
max_n_neighbors = 30
para_n_neighbors = [i+1 for i in range(max_n_neighbors)]
train_accuracy=[]
test_accuracy=[]

for n_neighbors in para_n_neighbors:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(df_train_x_scaled, df_train_y)
    train_accuracy.append(clf.score(df_train_x_scaled, df_train_y))
    test_accuracy.append(clf.score(df_test_x_scaled, df_test_y))
    
ds_neighbors = pd.DataFrame()
ds_neighbors["Neighbors"] = para_n_neighbors
ds_neighbors["TrainAccuracy"] = train_accuracy
ds_neighbors["TestAccuracy"] = test_accuracy
ds_neighbors
```

### 3. 결론 도출

```python
ds_neighbors.set_index("Neighbors", inplace = False).plot()
```

- neighbor = 6 일 때, 정확도가 가장 높은 것을 확인할 수 있다.

```python
clf_model = KNeighborsClassifier(n_neighbors=6)
clf_model.fit(df_train_x_scaled, df_train_y)
print("train data accuracy: {0: .3f}".format(clf_model.score(df_train_x_scaled, df_train_y)))
print("test data accuracy: {0: .3f}".format(clf_model.score(df_test_x_scaled, df_test_y)))
```

## 나이브베이지안 실습 p.253

### 1. 데이터 구성하기

```python
ds_numeric = ds_tc.select_dtypes(["int64","float64"])
```

```python
from sklearn.model_selection import train_test_split


ds_tc_y = ds_numeric["CHURN"]
ds_tc_x = ds_numeric.drop("CHURN", axis = 1, inplace = False)

ds_tc_train_x, ds_tc_test_x, ds_tc_train_y, ds_tc_test_y =\
train_test_split(ds_tc_x, ds_tc_y, test_size=0.30, random_state = 1234)

print("train data X size:{}".format(ds_tc_train_x.shape))
print("train data Y size:{}".format(ds_tc_train_y.shape))
print("test data X size:{}".format(ds_tc_test_x.shape))
print("test data Y size:{}".format(ds_tc_test_y.shape))
```

### 2. 모델 생성

```python
from sklearn.naive_bayes import GaussianNB

naive_model = GaussianNB()
naive_model.fit(X = ds_tc_train_x, y = ds_tc_train_y)
```

```python
naive_model.class_prior_.round(3)
```

- P(Y=0) = 0.791, P(Y=1) = 0.209

```python
print("Accuracy: {0: .3f}".format(naive_model.score(ds_tc_test_x,ds_tc_test_y)))
```

### 결론 도출

- P(Y=0) = 0.791, P(Y=1) = 0.209 이다.
- 정확도는 0.835로 괜찮은 성능을 가진 모델을 생성하였다.
