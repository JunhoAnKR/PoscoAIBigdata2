## 판매 데이터에서 결측치 채워넣기

```python
fill_median_func = lambda g: g.fillna(g.median())
```

```python
sales_data_fill_na = sales_data.groupby('CustomerID').apply(fill_median_func)
```

```python
sales_data_fill_na =  sales_data_fill_na.drop("CustomerID", axis=1).reset_index(0).reset_index(drop=True)
sales_data_fill_na.head()
```

```python
sales_data_fill_na.loc[user_data[sales_data['period'].isnull()].index, : ].head()
```
