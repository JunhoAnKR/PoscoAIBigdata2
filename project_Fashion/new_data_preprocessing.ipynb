```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
user_data = pd.read_csv("user_new.csv",engine = 'python',encoding='euc-kr')
user_data.head()
```

```python
# 결측치 확인
user_data.isnull().sum(axis=0)
```

```python
# 각 변수의
for i in user_data.columns:
    if user_data[i].dtype == 'O':
        counts = user_data[i].value_counts()
        print(counts)
```

```python
# age가 미상인 친구들 찾기
user_data.loc[user_data[user_data['age'] == '미상'].index, :]
```

```python
# exchange 마이너스값 0으로 변환
user_data["exchange"]=user_data["exchange"].abs()
user_data.head()
```

```python
# 데이터 불러오기2
sales_data = pd.read_csv("cloth-sales-history_new.csv",engine = 'python')
sales_data.head()
```

```python
# 결측치 확인
sales_data.isnull().sum(axis=0)
```

```python
counts_list = []
for i in sales_data.columns:
    if sales_data[i].dtype == 'O':
        counts = sales_data[i].value_counts()
        counts_list.append(counts)
counts_list
```

```python
func = lambda g:g.split(", ")[1].split(")")[0]
sales_data["period"] = sales_data.period.apply(func)
```
