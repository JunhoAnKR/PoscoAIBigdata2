```python
df_train = pd.DataFrame(train_set, index=range(len(train_set)))
df_test = pd.DataFrame(test_set, index=range(len(test_set)))
```

```python
df_train = df_train.drop("CustomerID", axis=1)
df_test = df_test.drop("CustomerID", axis=1)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.formula.api as smf
```

```python
formula_model = smf.ols(formula = "how_long ~ after_item"
                        , data = df_train)
formula_result = formula_model.fit()
print(formula_result.summary()) 
```

```python
df_test_x = df_test.drop("how_long", axis=1)
df_test_x
```

```python
import math
Y_pred = formula_result.predict(df_test_x)
test_y = df_test["how_long"]
```

```python
import matplotlib.font_manager as fm
import matplotlib
import matplotlib.pyplot as plt
for f in fm.fontManager.ttflist:
    if 'NanumBarunGothic' in f.fname: plt.rcParams['font.family']='NanumBarunGothic'
    # 폰트 나눔고딕으로

print(plt.rcParams['font.family'])

matplotlib.rcParams['axes.unicode_minus'] = False # '-'기호 제대로 나오게

plt.scatter(test_y, Y_pred)
plt.xlabel(u"실제 period")
plt.ylabel(u"period 예측치")
plt.title("period 예측치와 실제 period의 관계")
plt.show()
```

```python
regression = pd.DataFrame()
regression["actual_test_y"]= test_y
regression["pred_test_y"]= Y_pred
accuracy_list = []
for i in range(len(test_y)):
    if abs(test_y[i]-Y_pred[i]) <= 30:
        accuracy_list.append(1)
    else:
        accuracy_list.append(0)

regression.head()
```

```python
accuracy = sum(accuracy_list)/len(accuracy_list)
accuracy
```
