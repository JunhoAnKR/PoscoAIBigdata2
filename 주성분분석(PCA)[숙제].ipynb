고유벡터와 고유값의 의미를 잘 생각해보면, 

고유 벡터는 그 행렬이 벡터에 작용하는 힘의 방향을 나타내므로 공분산 행렬의 고유 벡터는 데이터가 어떤 방향으로 분산되어 있는지를 나타내준다고 할 수 있다.

고유 값은 고유벡터에 해당하는 상관 계수에 해당하므로, 우리가 다루는 행렬이 공분산 행렬일 경우 고유 값은 각 축에 대한 공분산 값이 된다. 따라서 고유 값이 큰 순서대로 고유 벡터를 정렬하면 결과적으로 중요한 순서대로 주성분을 구하는 것이 된다.

# PCA 실습 p.289

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy import stats
from sklearn.preprocessing import StandardScaler
```

### 1. 데이터 구성하기

```python
ds_iris = pd.read_csv("실습파일/IRIS.csv", engine = 'python')
ds_iris_x = ds_iris.drop("SPECIES", axis =1)
ds_iris_y = ds_iris["SPECIES"]
ds_iris_x.head()
```

```python
sns.pairplot(ds_iris_x)
```

### 2. 통계량 확인

```python
ds_iris.describe()
```

### 3. 상관관계 확인

```python
ds_iris_x.corr()
```

### 4. 상관계수 기반 주성분 실행

```python
scaler = StandardScaler()
ds_iris_x_scaled = scaler.fit_transform(ds_iris_x)

print("Scaled Data :\n{}".format(ds_iris_x_scaled[:5].round(3)))
print("Shape :\n{}".format(ds_iris_x_scaled.shape))
```

```python
pca = PCA(n_components = 4)
x_pca = pca.fit_transform(ds_iris_x_scaled)

print("PCA Data : \n{}".format(x_pca[:5]))
print("Shape : \n{}".format(x_pca.shape))

ds_x_pca = pd.DataFrame(x_pca, columns=["Prin1","Prin2","Prin3","Prin4" ])
```

### 5. 주성분, 기여율, Scree Plot확인, 주성분 저장 + 해석

```python
ds_pca = pd.DataFrame()
ds_pca["Prin"] = [ i for i in range(1, 5)]
ds_pca["EigenValues"] = pca.explained_variance_
ds_pca["EigenValueRatio"] = pca.explained_variance_ratio_
ds_pca["CumsumEigenValueRatio"] = np.cumsum(pca.explained_variance_ratio_)
ds_pca.round(3)
```

```python
ds_feature = pd.DataFrame({"Feature": ds_iris_x.columns})

ds_eigenvectors = pd.concat([ds_feature,
                            pd.DataFrame(pca.components_, columns=["Prin1","Prin2","Prin3","Prin4"])],
                           axis= 1)
ds_eigenvectors
```

```python
plt.plot(ds_pca["Prin"], ds_pca["EigenValueRatio"], label = "Ratio")

plt.plot(ds_pca["Prin"], ds_pca["CumsumEigenValueRatio"], label = "Cumsum Ratio")

plt.ylabel("Eigen Value ratio")
plt.xlabel("Prin")
plt.legend(loc="best")
```

- 주성분 분석 결과, 2개의 주성분 만으로 데이터 변동의 95.8%를 설명함을 알 수 있다.
- 따라서 제1주성분부터 제2주성분까지 주성분 2개를 사용함으로써 차원을 4개에서 2개로 줄여 시각화와 통계분석에 용이해졌다.

# PCA 숙제

### 1. 데이터 불러오기

```python
df = pd.read_csv("실습파일/유방암.csv", engine = 'python')
df.head()
```

### 2. 상관계수 확인

```python
df_x = df.drop("diagnosis", axis =1)
df_y = df["diagnosis"]
df_x.corr()
```

- 상관계수가 높은 변수들이 존재한다.

### 3. PCA 실행

```python
scaler = StandardScaler()
df_x_scaled = scaler.fit_transform(df_x)

print("Scaled Data :\n{}".format(df_x_scaled[:5].round(3)))
print("Shape :\n{}".format(df_x_scaled.shape))
```

```python
pca = PCA(n_components = 30)
x_pca = pca.fit_transform(df_x_scaled)

print("PCA Data : \n{}".format(x_pca[:5]))
print("Shape : \n{}".format(x_pca.shape))

col_names = []
for i in range(1,31):
    text = "Prin"
    text = text +str(i)
    col_names.append(text)


ds_x_pca = pd.DataFrame(x_pca, columns=col_names)
```

```python
ds_pca = pd.DataFrame()
ds_pca["Prin"] = [ i for i in range(1, 31)]
ds_pca["EigenValues"] = pca.explained_variance_
ds_pca["EigenValueRatio"] = pca.explained_variance_ratio_
ds_pca["CumsumEigenValueRatio"] = np.cumsum(pca.explained_variance_ratio_)
ds_pca.round(3)
```

```python
ds_feature = pd.DataFrame({"Feature": df_x.columns})

ds_eigenvectors = pd.concat([ds_feature,
                            pd.DataFrame(pca.components_, columns=col_names)],
                           axis= 1)
ds_eigenvectors
```

```python
plt.plot(ds_pca["Prin"], ds_pca["EigenValueRatio"], label = "Ratio")

plt.plot(ds_pca["Prin"], ds_pca["CumsumEigenValueRatio"], label = "Cumsum Ratio")

plt.ylabel("Eigen Value ratio")
plt.xlabel("Prin")
plt.legend(loc="best")
```

- 주성분 분석 결과, 5개의 주성분 만으로 데이터 변동의 84.7%를 설명함을 알 수 있다.
- 따라서 제 1주성분부터 제 5성분까지 5개의 성분을 사용하기로 결정한다.

### 4. 그림 그리기

```python
df_pca_for_scatter =pd.DataFrame()
df_pca_for_scatter["diagnosis"] = df_y
df_pca_for_scatter_all = pd.merge(df_pca_for_scatter, ds_x_pca.iloc[:,0:5], how='outer', left_index=True,right_index=True)
df_pca_for_scatter_all.head()
```

```python
sns.pairplot(df_pca_for_scatter_all, hue="diagnosis", markers=["o", "s"])
```

### 5. 결론 도출

- 주성분 분석 결과 데이터 변동의 80%이상을 설명하는 주성분 5개를 선택하였다.
- 또한 이들의 산점도를 그려 봤을 때, 히스토그램이 그려진 대각선이아닌 역 대각선 아래쪽(즉, 제 1,2 주성분과 그리지 않은 산점도들)을 보면 제 1,2 주성분을 포함하여 산점도를 그렸을 때 보다, diagnosis를 잘 분류하지 못하는 것을 알 수 있다.
- 이는 데이터 분산 설명력이 낮은 주성분이기 때문으로 생각된다.
