```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. 정규분포 그리기 : random data 생성 : n = 100000, xbar = 30, sd = 2.5
np.random.seed(seed= 1234)
normal = np.random.normal(30, 2.5, 100000)
sns.distplot(normal)
```

```python
# 2. 대한민국 남성의 20~30대 키를 측정한 결과, 평균 175cm, 표준편차 5cm, 180cm일 경우 몇 %??
normal4 = stats.norm(loc=175, scale=5)
xx = np.linspace(160, 190, 400)
plt.plot(xx,normal4.pdf(xx))
x=180
v3 = stats.norm.cdf(x, 175, 5)
print("{0}의 키는 상위 {1: .1f}%".format(x,(1-v3)*100))
```

```python
# 3.어떤 자동차 1리터당 주행거리가 평균 12cm, 표준편차 3cm인 정규분포를 이룬다고 하자.
normal5 = stats.norm(loc = 12, scale = 3)
xx = np.linspace(3, 21, 400)
plt.plot(xx, normal5.pdf(xx))
num1 = 12
num2 = 15
num3 = 9

v1 = stats.norm.cdf(num1, 12, 3)
v2 = stats.norm.cdf(num2, 12, 3)
v3 = stats.norm.cdf(num3, 12, 3)

print("1리터를 가지고 {0} 이상 {1} 이하를 달릴 확률은 : {2: .1f}".format(num1, num2, (v2-v1)*100))
print("1리터를 가지고 {0} 이상 {1} 이하를 달릴 확률은 : {2: .1f}".format(num3, num2, (v2-v3)*100))
print("1리터를 가지고 {0} 이상 달릴 확률은 : {1: .1f}".format(num2, (1-v2)*100))
```

```python
# 4. 표준정규분포 그리기
normal6 = stats.norm(0, 1)
xx= np.linspace(-3,3,400)
plt.plot(xx , normal6.pdf(xx))
```

```python
# 5. t-분포 그리기
np.random.seed(seed=1234)
xx = np.linspace(-4, 4, 400)
for df in [10, 50]:
    t_df = stats.t(df=df)
    plt.plot(xx, t_df.pdf(xx))
```

```python
# 6. 확률변수 t는 자유도가 5인 t-분포를 따른다. P(T <= t)를 계산하시오
t = 1.53
df = 5
prop = stats.t.cdf(t, df)
print("P(T<={0}): {1: .3f}".format(t,prop))
```

```python
# 7. 확률변수 t는 자유도가 10인 t-분포를 따른다. P(T <= t)를 계산하시오
t = 2.0
df = 10
prop = stats.t.cdf(t, df)
print("P(T<={0}) : {1: .3f}".format(t, prop))
```

```python
# 8. 카이제곱 분포 그리기
np.random.seed(seed=1234)
xx = np.linspace(0, 100, 400)
for df in [3, 50]:
    chi_df = stats.chi2(df=df)
    plt.plot(xx, chi_df.pdf(xx))
```

```python
# 9. Chisq 
np.random.seed(1234)
chisq = 3.0
df = 8

prop = stats.chi2.cdf(chisq, df)
print("P(T<={0}) : {1: .4f}".format(chisq, prop))
prop2 = stats.chi2.sf(chisq, df)
print("P(T>={0}) : {1: .4f}".format(chisq, prop2))
```

```python
# 10. F 분포
np.random.seed(1234)
f_df10 = np.random.f(dfnum = 10, dfden = 10, size=1000)
sns.distplot(f_df10, kde = True)

f_df50 = np.random.f(dfnum = 10, dfden = 50, size = 100)
sns.distplot(f_df50,kde = True)
```

```python
# 11. 와이블분포 
np.random.seed(1234)
x = 1200
alpha = 2.5
beta = 1500

prop = stats.weibull_min.cdf(x, c = alpha, scale = beta)
print("P(x>={0}) : {1: .5f}".format(x, (1-prop)))
```

```python
# 12. 이항 분포
n = 5
for i in range(n+1):
    prop = stats.binom.pmf(k = i, n=n, p = 0.3)
    print("P(X={0}) : {1: .3f}".format(i, prop))
```

```python
# 13. poisson
mu = 2
prop = stats.poisson.pmf(2, mu)
cdf_prop = stats.poisson.cdf(1, mu)
sf_prop = stats.poisson.sf(3, mu)

print("10시부터 11시 사이에{0}명의 손님이 도착 할 확률: {1: .4f}".format(2, prop))
print("10시부터 11시 사이에{0}명 이하의 손님이 도착 할 확률: {1: .4f}".format(1, cdf_prop))
print("10시부터 11시 사이에{0}명 이상의 손님이 도착 할 확률: {1: .4f}".format(3, sf_prop))
```

```python
# 14. 기술통계 : 자동차 연비 데이터 셋에서 기술통계치 구하기
# 14-1. 시내에서 연비 통계치 구하기: 모델별 데이터수, 평균, 중앙값, 표준편차, 최소값, 최대값, 사분위수
ds_mycars = pd.read_csv("실습파일/mycars.csv", engine = "python")
ds_mycars[['model', 'mpg']].groupby('model').describe()
```

```python
# 15.자동차의 모델과 연비에 대해 counts, percents, cumulative counts, cumulative percents를 계산하라.
df = ds_mycars[['model', 'mpg']]
count = df['model'].value_counts().sort_index()
cumcnt= np.cumsum(count)
percent = count/sum(count)*100
cumpct = np.cumsum(percent)
count_data = pd.DataFrame({'Count': count, 'CumCnt': cumcnt, 'Percent': percent,'CumPct': cumpct })
count_data
```

```python
# 16. 추론통계 - 모표준편차를 아는 경우
df = pd.DataFrame({'sample': [18, 18, 20, 21, 20, 23, 19, 18, 17, 21, 22, 20, 20, 21, 20, 19, 19, 18, 17, 19]})
lower, upper = stats.norm.interval(0.95, loc= df.mean(), scale =3.8/np.sqrt(len(df)))
print("신뢰구간: ({0}, {1})".format(lower.round(2), upper.round(2)))
```

```python
# 17. 추론통계 - 모표준편차를 모르는 경우
df = pd.DataFrame({'sample': [78,71,74,69,70,79,70,68,75,72,70,72,73,70,70,72,71,70,75,72]})
lower, upper = stats.t.interval(0.95, len(df)-1, loc= 71.50, scale =1.933/np.sqrt(len(df)))
print("신뢰구간: ({0}, {1})".format(lower.round(2), upper.round(2)))
```

```python
# 18. 1-sample t-test
df = pd.DataFrame({'sample' :[85.0,79.0,79.1,79.9,81.6,78.6,85.4,83.4,78.1,79.2]})
t_result = stats.ttest_1samp(df, 76.7)
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)
print("1-sample t-test")
print("t검정통계량: {}".format(t))
print("p-value : {}".format(p))
## p-value < 0.05 이므로 개선활동이 만족도를 변화 시켰다고 할 수 있다.
```

```python
# 19. paired t-test
df1 = pd.DataFrame({'before' : [720, 589, 780, 648, 720, 589, 780, 648, 780, 648]})
df2 = pd.DataFrame({'after'  : [710, 580, 787, 712, 750, 600, 782, 670, 790, 680]})
t_test = stats.ttest_rel(df1, df2)
t , p = t_test.statistic.round(3), t_test.pvalue.round(3)
print("1-sample t-test")
print("t검정통계량: {}".format(t))
print("p-value : {}".format(p))
# P-value !<=0.05 이므로 회사원들의 성적은 학원가기 전후에 대해 차이가 없다고 할 수 있다.
```

```python
# 20. 1 proportion test
from statsmodels.stats.proportion import proportions_ztest
count = 10
nobs = 100
value = 0.2

stat, pval = proportions_ztest(count, nobs, value)
print("1 Proportion z-test")
print("p검정통계량: {0:0.3f}".format(stat))
print("p-value : {0:0.3f}".format(pval))
# p-value < 0.05 이므로 품질개선 결과, 기존보다 전체 고객 중 만족도의 차이가 있다고 할 수 있다.
```

```python
# 21. 2-proportion test
count = np.array([14, 5])
nobs = np.array([1200, 1200])

stat, pval = proportions_ztest(count, nobs)
print("1 Proportion z-test")
print("p검정통계량: {0:0.3f}".format(stat))
print("p-value : {0:0.3f}".format(pval))
# p-value < 0.05 이므로 두 공정의 불량률이 같다고 할 수 있다.
```

```python
# 22. chi-square test
df = pd.DataFrame({'A/S':[18,8,4,4,3,3],
                   '가격':[1,2,1,1,1,25],
                   '성능':[8,14,3,2,3,8],
                   '확장성':[7,5,4,3,1,10],
                   '디자인':[10,5,9,2,1,2],
                   '안정성':[9,9,5,7,1,1],
                   '기능성':[10,4,4,3,1,7]})
chi, pval, dof, expected = stats.chi2_contingency(df.T)
print("Chi-square test")
print("chisq: {0:0.3f}".format(chi))
print("p: {0:0.3f}".format(pval))
print("degree of freedom: {}".format(dof))
print("expected value:: \n{}".format(expected.round(3)))
# p-value < 0.05 이므로 각 제품에 대해 이미지에 대해 차이가 있다고 할 수 있다.
```

```python
# 1. Correlation Analysis
Strength = [37.6, 38.6, 37.2, 36.4, 38.6, 39, 37.2, 36.1, 35.9, 37.1, 36.9, 37.5, 36.3, 38.1, 39, 36.9, 36.8, 37.6
       , 33, 33.5]
Temperature = [14, 15, 14, 16, 17, 14, 17, 16, 15, 16, 14, 16, 18, 13, 15, 17, 14, 16, 20, 21]
plt.scatter(Temperature,Strength)
corr, pval = stats.pearsonr(Strength, Temperature)
print("Correlation Analysis")
print("corr: {0: .3f}".format(corr))
print("p: {0: .3f}".format(pval))

# p<0.05 이므로 df1, df2는 상관관계가 있다고 할 수 있다.
```

```python
# 2 . Simple Linier Regression

Strength = [37.6, 38.6, 37.2, 36.4, 38.6, 39, 37.2, 36.1, 35.9, 37.1, 36.9, 37.5, 36.3, 38.1, 39, 36.9, 36.8, 37.6
       , 33, 33.5]
BakingTime = [0.798922, 0.84902, 0.816163, 0.758266, 0.815894, 0.807477, 0.809068, 0.7664, 0.742243, 0.751118
      ,0.740629, 0.751495, 0.738863, 0.827428, 0.818172, 0.740939, 0.75728, 0.804695, 0.695121, 0.735377]

BakingTime = sm.add_constant(BakingTime) # X 

model = sm.OLS(Strength, BakingTime) # y, x
result = model.fit()

print(result.summary())
```
