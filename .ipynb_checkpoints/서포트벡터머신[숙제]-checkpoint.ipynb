## 서포트벡터머신_실습 p.221

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.svm import SVC
```

### 1. 데이터 구성하기

```python
ds_tc = pd.read_csv("실습파일/통신고객이탈.csv")
ds_tc.head()
```

```python
ds_tc["CHURN"].replace(to_replace = "Active", value = 0, inplace = True)
ds_tc["CHURN"].replace(to_replace = "Churned", value = 1, inplace = True)
```

```python
ds_tc.isnull().sum(axis=0)
```

```python
ds_tc_dummy = pd.get_dummies(ds_tc.drop(["CHURN", "CUSTOMER_ID"], axis =1, inplace = False))
ds_tc_dummy.head()
```

```python
from sklearn.model_selection import train_test_split


ds_tc_y = ds_tc["CHURN"]
ds_tc_x = ds_tc_dummy

ds_tc_train_x, ds_tc_test_x, ds_tc_train_y, ds_tc_test_y =\
train_test_split(ds_tc_x, ds_tc_y, test_size=0.30, random_state = 1234)

print("train data X size:{}".format(ds_tc_train_x.shape))
print("train data Y size:{}".format(ds_tc_train_y.shape))
print("test data X size:{}".format(ds_tc_test_x.shape))
print("test data Y size:{}".format(ds_tc_test_y.shape))
```

```python
# 스케일에 영향을 많이 받는 SVM이므로 스케일링한다.
ds_tc_train_x_scaled = scale(ds_tc_train_x,axis = 0)
ds_tc_test_x_scaled = scale(ds_tc_test_x,axis = 0)
pd.DataFrame(ds_tc_train_x_scaled, columns=ds_tc_train_x.columns).head()
```

### 2. 모델 생성 - Parameter 조정 / 평가

```python
svm_uncustomized_scaled = SVC(random_state=1234)
svm_uncustomized_scaled.fit(ds_tc_train_x_scaled, ds_tc_train_y)

print("Accuracy on training set: {:.3f}".format(svm_uncustomized_scaled.score(ds_tc_train_x_scaled, ds_tc_train_y)))
print("Accuracy on test set: {:.3f}".format(svm_uncustomized_scaled.score(ds_tc_test_x_scaled, ds_tc_test_y)))
```

```python
train_accuracy = []
test_accuracy = []
para_C = []

for C in range(15):
    svm1_scaled = SVC(C = (C+1)/10, random_state=1234)
    svm1_scaled.fit(ds_tc_train_x_scaled, ds_tc_train_y)
    para_C.append((C+1)/10)
    train_accuracy.append(svm1_scaled.score(ds_tc_train_x_scaled, ds_tc_train_y))
    test_accuracy.append(svm1_scaled.score(ds_tc_test_x_scaled, ds_tc_test_y))
    
ds_accuracy1_scaled = pd.DataFrame()
ds_accuracy1_scaled["C"] = para_C
ds_accuracy1_scaled["TrainAccuracy"] = train_accuracy
ds_accuracy1_scaled["TestAccuracy"] = test_accuracy
```

```python
ds_accuracy1_scaled.round(3)
```

```python
ds_accuracy1_scaled.set_index("C", inplace=False).plot.line()
```

```python
train_accuracy = []
test_accuracy = []
para_gamma = []

for gamma in range(1,10):
    svm2_scaled = SVC(gamma=gamma/10, random_state=1234)
    svm2_scaled.fit(ds_tc_train_x_scaled, ds_tc_train_y)
    para_gamma.append(gamma/10)
    train_accuracy.append(svm2_scaled.score(ds_tc_train_x_scaled, ds_tc_train_y))
    test_accuracy.append(svm2_scaled.score(ds_tc_test_x_scaled, ds_tc_test_y))
    
ds_accuracy2_scaled = pd.DataFrame()
ds_accuracy2_scaled["Gamma"] = para_gamma
ds_accuracy2_scaled["TrainAccuracy"] = train_accuracy
ds_accuracy2_scaled["TestAccuracy"] = test_accuracy
```

```python
ds_accuracy2_scaled.round(3)
```

```python
ds_accuracy2_scaled.set_index("Gamma", inplace=False).plot.line()
```

```python
train_accuracy = []
test_accuracy = []

svm_model1 = SVC(C=0.3, gamma=0.1, random_state=1234)
svm_model2 = SVC(C=0.5, gamma=0.1, random_state=1234)
svm_model3 = SVC(C=0.7, gamma=0.1, random_state=1234)

train_accuracy.append(svm_model1.fit(ds_tc_train_x_scaled, ds_tc_train_y).score(ds_tc_train_x_scaled, ds_tc_train_y))
train_accuracy.append(svm_model2.fit(ds_tc_train_x_scaled, ds_tc_train_y).score(ds_tc_train_x_scaled, ds_tc_train_y))
train_accuracy.append(svm_model3.fit(ds_tc_train_x_scaled, ds_tc_train_y).score(ds_tc_train_x_scaled, ds_tc_train_y))

test_accuracy.append(svm_model1.fit(ds_tc_train_x_scaled, ds_tc_train_y).score(ds_tc_test_x_scaled, ds_tc_test_y))
test_accuracy.append(svm_model2.fit(ds_tc_train_x_scaled, ds_tc_train_y).score(ds_tc_test_x_scaled, ds_tc_test_y))
test_accuracy.append(svm_model3.fit(ds_tc_train_x_scaled, ds_tc_train_y).score(ds_tc_test_x_scaled, ds_tc_test_y))
    
ds_accuracy2 = pd.DataFrame(index = ["svm_model1", "svm_model2", "svm_model3"])
ds_accuracy2["TrainAccuracy"] = train_accuracy
ds_accuracy2["TestAccuracy"] = test_accuracy
ds_accuracy2
```

### 3. 결론 도출

```python
svm_model3.get_params
```

```python
svm_model3.support_vectors_
```

- 모델 평가 결과,C = 0.7, gamma = 0.1일 때, 정확도가 가장 높게 나왔다.

## 서포트벡터머신_숙제

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

### 1. 데이터 불러오기 

```python
df = pd.read_csv("실습파일/유방암.csv", engine = 'python')
df.head()
```

```python
df_dummy = pd.get_dummies(df)
df_dummy.head()
```

```python
from sklearn.model_selection import train_test_split


df_y = df["diagnosis"]
df_x = df_dummy.drop("diagnosis", axis = 1, inplace = False)

df_train_x, df_test_x, df_train_y, df_test_y =\
train_test_split(df_x, df_y, test_size=0.30, random_state = 1234)


print("train data X size:{}".format(df_train_x.shape))
print("train data Y size:{}".format(df_train_y.shape))
print("test data X size:{}".format(df_test_x.shape))
print("test data Y size:{}".format(df_test_y.shape))
```

```python
# 스케일에 영향을 많이 받는 SVM이므로 스케일링한다.
df_train_x_scaled = scale(df_train_x,axis = 0)
df_test_x_scaled = scale(df_test_x,axis = 0)
pd.DataFrame(df_train_x_scaled, columns=df_train_x.columns).head()
```

### 2. 모델 생성하기

```python
svm_uncustomized_scaled = SVC(random_state=1234)
svm_uncustomized_scaled.fit(df_train_x, df_train_y)

print("Accuracy on training set: {:.3f}".format(svm_uncustomized_scaled.score(df_train_x, df_train_y)))
print("Accuracy on test set: {:.3f}".format(svm_uncustomized_scaled.score(df_test_x, df_test_y)))
```

```python
svm_uncustomized_scaled = SVC(random_state=1234)
svm_uncustomized_scaled.fit(df_train_x_scaled, df_train_y)

print("Accuracy on training set: {:.3f}".format(svm_uncustomized_scaled.score(df_train_x_scaled, df_train_y)))
print("Accuracy on test set: {:.3f}".format(svm_uncustomized_scaled.score(df_test_x_scaled, df_test_y)))
```

```python
# explore best parameter
parameters = {'C':np.arange(0.0,2,0.1)[1:],'gamma':np.arange(0.0,2,0.1)[1:]} 

clfr = SVC(random_state=1234, kernel='rbf', probability=True)

grid = GridSearchCV(clfr, parameters,scoring='accuracy', cv=8) 
grid.fit(df_train_x_scaled, df_train_y) 

print('The parameters combination that would give best accuracy is : ') 
print(grid.best_params_) 
print('The best accuracy achieved after parameter tuning via grid search is : ', grid.best_score_) 
```

- The parameters combination that would give best accuracy is : {'gamma': 0.1, 'C': 1.5}

```python
grid.fit(df_train_x_scaled, df_train_y).score(df_test_x_scaled, df_test_y) 
```

- Accuracy of test data set is about 92%

```python
import matplotlib.font_manager as fm
import matplotlib
import matplotlib.pyplot as plt
for f in fm.fontManager.ttflist:
    if 'NanumBarunGothic' in f.fname: plt.rcParams['font.family']='NanumBarunGothic'
    # 폰트 나눔고딕으로

print(plt.rcParams['font.family'])

matplotlib.rcParams['axes.unicode_minus'] = False # '-'기호 제대로 나오게
```

```python
from sklearn.metrics import roc_curve, auc

#svm_model = SVC(C= 1.5, gamma = 0.1,kernel='rbf', random_state=1234)
svm_model = grid.fit(df_train_x_scaled, df_train_y)
```

```python
fpr, tpr, thresholds = roc_curve(df_test_y, svm_model.predict_proba(df_test_x_scaled)[:, 1])
auc_score = auc(fpr, tpr).round(3)
plt.plot(fpr, tpr, 'o-', ms=2, label="Kernel SVM(AUC = %s)"%auc_score)
plt.legend()
plt.plot([0, 1], [0, 1], 'k--', label="random guess")
plt.xlabel('False Positive Rate (Fall-Out)')
plt.ylabel('True Positive Rate (Recall)')
plt.title('ROC 커브')
plt.show()
```

- X축 : 1-특이도, Y축 : 민감도
- 민감도는 질병이 실제로 있는데 Test 결과에서도 질병이 있다고 판단하는 비율
- 특이도는 질병이 실제로 없는데 Test 결과도 질병이 없다고 판단하는 비율
- ROC curve : 특이도가 감소하는 속도에비해 민감도가 얼마나 빠르게 증가하는지를 나타냄.
- AUC = AUROC (the Area Under a ROC Curve) : ROC 커브의 밑면적을 구한 값이 바로 AUC. 이 값이 1에 가까울수록 성능이 좋다.
- 그러나 어느쪽에 좀 더 강조를 둘것인가가가 중요할 수 있다. 예를 들어 걸릴 확률은 매우 낮지만, 치사율이 극히 높은 병은 민감도가 높아야하고, 걸릴 확률은 높지만, 위험성이 거의 없는 경우는 민감도가 낮아도 괜찮을 것이다.)
