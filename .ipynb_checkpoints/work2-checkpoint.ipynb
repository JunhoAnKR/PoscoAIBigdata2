```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = [1, 2, 3, 4, 4, 5, 6, 6, 7, 8]
df2 = [23, 29, 49, 64, 74, 87, 96, 97, 109, 119]

corr, pval = stats.pearsonr(df1, df2)
print("Correlation Analysis")
print("corr: {0: .3f}".format(corr))
print("p: {0: .3f}".format(pval))
```

```python
pd.concat([df1,df2,df3])
```

```python
# concat 활용
df4 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},index = [0,1,2,3])
df5 = pd.DataFrame({'E':['E0','E1','E2','E3'],
                    'F':['F0','F1','F2','F3'],
                    'G':['G0','G1','G2','G3'],
                    'H':['H0','H1','H2','H3']},index = [0,1,2,3])
df6 = pd.DataFrame({'I':['I0','I1','I2','I3'],
                    'G':['G0','G1','G2','G3'],
                    'K':['K0','K1','K2','K3'],
                    'L':['L0','L1','L2','L3']},index = [0,1,2,3])
result = pd.concat([df4,df5,df6],axis=1)
result 
```

```python
left = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                     'key2':['K0','K1','K0','K1'],
                     'A'   :['A0','A1','A2','A3'],
                     'B'   :['B0','B1','B2','B3']})

right = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                      'key2':['K0','K0','K0','K0'],
                      'C'   :['C0','C1','C2','C3'],
                      'D'   :['D0','D1','D2','D3']})
```

```python
pd.merge(left,right, on=['key1', 'key2'], how='inner')
```

```python
pd.merge(left,right, on=['key1', 'key2'], how='outer')
```

```python
left = pd.read_csv('실습파일/FITNESS_1.csv', engine = 'python', encoding ='EUC-KR')

right = pd.read_csv('실습파일/FITNESS_2.csv', engine = 'python', encoding ='EUC-KR')
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='NAME',how='outer')
```

```python
df_merge =pd.merge(left, right, on=['NAME','GENDER','AGE'])
```

```python
df_merge[df_merge['GENDER']=='여성']
```

```python
df_merge.groupby("GENDER").mean()
```

```python
df_merge.groupby(["GENDER","AGEGROUP"]).mean()[['WEIGHT']]
```

```python
df_merge.groupby(["GENDER","AGEGROUP"])['WEIGHT'].agg(['mean', 'std', 'median', 'min','max'])
```

```python
df_merge.groupby(['GENDER', 'AGEGROUP'])['WEIGHT', 'OXY', 'RUNTIME'].agg({'WEIGHT': 'mean', 'OXY':'std','RUNTIME':'max'})
```

```python
df = pd.DataFrame({'foo' : ['one','one','one','two','two','two'],
                   'bar' : ['A','A','C','A','B','C'],
                   'baz' : [1,2,3,4,5,6],
                   'zoo' : ['x','y','z','q','w','t']})
df
```

```python
df.pivot_table(index = 'foo', columns = 'bar', values = 'baz', aggfunc = sum) # aggfunc = 중복 처리함수
```

```python
df.melt(id_vars = ['foo','bar'])
```

```python
df.melt(id_vars = ['foo','bar'], value_vars = ['zoo'])
```
